```julia
using LinearAlgebra,Random,Printf,FFTW,JLD,Plots,LaTeXStrings,Colors, DelimitedFiles
include("2.1_Markov_chain.jl");
```

```julia
N = 100
D = 2
save = true
Î², b = 0.43, 0.01
seed = 1
N_config = 200

M, E = Markov_chain(Î², b, N, D,"hot" , N_config)


x = collect(0:N_config)
p = scatter(x, M, color = 1, label = "M(s)"); scatter!(x, E, color = 2, label = "E(s)")
xlabel!("Configuration number")
display(p)
if save
    F = hcat(x, M, E)
    writedlm(string("D", D ,"_N", N, "_",seed,".txt"), F)
end
```
