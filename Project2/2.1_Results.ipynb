```julia
using LinearAlgebra,Random,Printf,FFTW,JLD,Plots,LaTeXStrings,Colors
include("2.1_Markov_chain.jl")
```

```julia
N = 100
D = 2

Î², b = 0.43, 0.01
seed = 12345
N_config = 200

s = s_init(N, D, "hot")
M, E = Markov_chain(Î², b, s, seed, N_config)

x = collect(0:N_config)
p = scatter(x, M, color = 1, label = "M(s)"); scatter!(x, E, color = 2, label = "E(s)")
xlabel!("Configuration number")
display(p)
```

```julia
cutoff = 40

println(sum(M[cutoff:end])/length(M[cutoff:end]))
println(sum(E[cutoff:end])/length(E[cutoff:end]))

h = histogram(E[cutoff:end], bins = 30, label = "")
display(h)
```
